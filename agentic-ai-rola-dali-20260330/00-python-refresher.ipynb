```python
class SimpleAgent:
    def __init__(self, name: str) -> None:
        self.name = name

    def respond(self, message) -> str:
        return f"{self.name} received {message}"


class CallableAgent:
    def __init__(self, name: str) -> None:
        self.name = name

    def __call__(self, message: str) -> str:
        return f"{self.name} received {message}"
```

```python
simpleAgent = SimpleAgent("Agent A")
simpleAgent.respond("Without Call")

callableAgent = CallableAgent("Agent C")
callableAgent("this is a message to the callable agent")
```
